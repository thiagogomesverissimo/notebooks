```python
import numpy as np
import pandas as pd

pd.set_option('display.max_columns',100) # VER todas colunas

df = pd.read_csv('data/nhanes_2015_2016.csv')
df.head()
```

```python
# Criar duas novas colunas com o min e max das colunas:
# BPXSY1 BPXDI1 BPXSY2 BPXDI2 BMXWT BMXHT BMXBMI BMXLEG BMXARML BMXARMC BMXWAIST
cols=['BPXSY1','BPXDI1','BPXSY2','BPXDI2','BMXWT','BMXHT','BMXBMI','BMXLEG','BMXARML','BMXARMC','BMXWAIST']

def min_max(row):
    data = row[cols]
    return pd.Series({'min': np.min(data), 'max': np.max(data)})

df.apply(min_max,axis='columns').head()
```

```python
def min_max(row):
    data = row[cols]
    row['min'] = np.min(data)
    row['max'] = np.max(data)
    return row

df.apply(min_max,axis='columns').head()
```
