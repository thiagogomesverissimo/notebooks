```python
import numpy as np
from PIL import Image
from IPython.display import display
```

```python
thiago = Image.open('data/thiago.tif')
display(thiago)
```

```python
array = np.array(thiago)
print(array.shape)
print(array.dtype)
```

```python
# 0 - preto
# 255 - branco
mask = np.full(array.shape,255)
mask
```

```python
modificado = (mask-array)
modificado = modificado.astype(np.uint8)
modificado
```

```python
display(Image.fromarray(modificado))
```

```python
reshaped = np.reshape(modificado,(200,800))
display(Image.fromarray(reshaped))
```
