Análise das contratações de analistas de sistemas da USP usando como referência dados do https://uspdigital.usp.br/portaltransparencia/ coletados em setembro de 2024.

Obervação: programando por diversão, copiem os códigos e façam suas próprias análises

```python
import pandas as pd
import matplotlib.pyplot as plt

# txt exportado do portal de transparência da USP
df = pd.read_csv('../data/2024/usp-analistas-de-sistemas-set-2024.txt',sep=';', encoding='latin-1',index_col=16)
```

```python
# filtrando apenas Celetistas e Autarquicos (Func Aut)
df = df[df['Categoria'].isin(['Celetista','Func Aut'])]
```

```python
# ingressantes em 2024
df = df[df['Data Ingresso/Aposentadoria'].str.contains('2024')]
df.shape[0]
```

Foram contratados 32 Analistas de Sistema na USP em 2024

```python
todas_funcoes = df.groupby(['Unid/Orgão']).size()
todas_funcoes.to_frame('size')
```

```python
ax = todas_funcoes.plot(kind='bar')
ax.bar_label(ax.containers[0])
plt.title("Quantidade de contratações de Analistas de Sistema USP 2024")
plt.ylabel("Quantidade de pessoas")
plt.savefig('/tmp/analistas.png')
```
