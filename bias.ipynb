```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

Vamos considerar que nossa população hipotética são habitantes da cidade de são paulo e nossa variável de interesse é a **idade** das pessoas. Vamos considerar dois subgrupos, um de universitários e outro não. Vamos supor que 15% da população são universitários:

```python
random.seed(1234)
tamanho_populacao = 100000

nao_universitarios_media = 50
nao_universitarios_sd = 10

universitarios_media = 20
universitarios_sd = 3

porcentagem_universitarios = .15
```

```python
universitarios = np.random.normal(universitarios_media,
                                      universitarios_sd,
                                      int(tamanho_populacao*porcentagem_universitarios))

nao_universitarios = np.random.normal(nao_universitarios_media,
                                      nao_universitarios_sd,
                                      int(tamanho_populacao*(1-porcentagem_universitarios)))

populacao = np.append(universitarios,nao_universitarios)

# Gráficos

sns.displot(universitarios)
plt.title("Universiários")
plt.xlim([10,80])

sns.displot(nao_universitarios)
plt.title("Não Universitários")
plt.xlim([10,80])

sns.displot(populacao)
plt.title("População")
plt.axvline(x = np.mean(populacao))
plt.xlim([10,80])

plt.show()
```

Vamos supor que faremos um experimento na qual selecionaremos 500 amostras diretamente da população com com 50 observações em cada amostra:

```python
amostras = 500
tamanho_cada_amostra = 50

medias_das_amostras = np.empty(amostras)
for i in range(amostras):
    random_populacao = np.random.choice(populacao, tamanho_cada_amostra)
    medias_das_amostras[i] = np.mean(random_populacao)
    
sns.displot(populacao)
plt.title("População")
plt.axvline(x = np.mean(populacao))
plt.xlim([10,80])

sns.displot(medias_das_amostras)
plt.title("Disrtribuição da amostragem usando toda população")
plt.axvline(x = np.mean(populacao))
plt.axvline(x = np.mean(medias_das_amostras), color = "black")
plt.xlim([10,80])

plt.show()
```

Agora criaremos um viés fazendo uma amostragem não-representativa e usaremos apenas da população de universitários:

```python
amostras = 500
tamanho_cada_amostra = 5

medias_das_amostras = np.empty(amostras)
for i in range(amostras):
    random_populacao = np.random.choice(universitarios, tamanho_cada_amostra)
    medias_das_amostras[i] = np.mean(random_populacao)
    
sns.displot(populacao)
plt.title("População")
plt.axvline(x = np.mean(populacao))
plt.xlim([10,80])

sns.displot(medias_das_amostras)
plt.title("Disrtribuição da amostragem enviesada com universitários")
plt.axvline(x = np.mean(populacao))
plt.axvline(x = np.mean(universitarios), color = "black")
plt.xlim([10,80])

plt.show()
```
