```python
from PIL import Image
import matplotlib as mtplb
import numpy as np
import matplotlib.pyplot as plt
```

```python
# https://stackoverflow.com/questions/7391945/how-do-i-read-image-data-from-a-url-in-python
import shutil
import requests

image_url = 'https://thiagogomesverissimo.github.io/images/thiago.jpg'
response = requests.get(image_url, stream=True)
with open('thiago.jpg', 'wb') as file:
    shutil.copyfileobj(response.raw, file)
del response
```

```python
im = Image.open("thiago.jpg").convert('L')
im.show()
```

```python
histograma = np.array(im.histogram())
plt.ylim([0, 30000])
plt.plot(histograma,'.')
```
