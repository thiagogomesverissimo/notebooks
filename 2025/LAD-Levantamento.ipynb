```python
import pandas as pd

pd.set_option('display.max_colwidth', None) # mostrar todo dataframe sempre
```

```python
data = pd.read_csv('../data/2025/lad-levantanento/articles.csv')
#data.columns
```

### Status atual da classificação de cada artigo, terminaremos quando os Unclassified chegar em zero

```python
pd.DataFrame(data.groupby('status').size(), columns=['quantidade'])
```

### Quantidade de artigos avaliados por cada membro da equipe

```python
pd.DataFrame(data.groupby('updated_by').size(), columns=['quantidade'])
```

### Artigos agrupados por status e critérios de seleção

```python
pd.DataFrame(data.groupby(['selection_criteria','status']).size(), columns=['quantidade'])
```

### Artigos que foram aceitos e possuem código aberto

```python
c1=data['selection_criteria']=='O Artigo apresenta um LAD e disponibiliza um link para o código fonte' 
c2=data['status']=='Accepted'
articles_with_code = data[c1 & c2]
articles_with_code[['updated_by','year','journal','title','author']]
```

### Artigos que foram aceitos e não possuem código aberto, mas possuem imagens

```python
c1=data['selection_criteria']=='O Artigo apresenta um LAD com imagens, mas não disponiliza o código' 
c2=data['status']=='Accepted'
articles_with_images = data[c1 & c2]
articles_with_images[['updated_by','year','journal','title','author']]
```

### Artigos aceitos com comentários

```python
c1=data['comments'].notna()
c2=data['status']=='Accepted'
articles_with_comments = data[c1 & c2]
articles_with_comments[['updated_by','title','status','comments']]
```

### Artigos rejeitados com comentários

```python
c1=data['comments'].notna()
c2=data['status']=='Rejected'
articles_with_comments = data[c1 & c2]
articles_with_comments[['updated_by','title','status','comments']]
```
