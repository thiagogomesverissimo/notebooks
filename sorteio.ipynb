## Problema do sorteio

*Dados de entrada:* 
 - Uma tabela de N avaliadores e a sigla dos seus respectivos departamento
 - Uma tabela de M avaliados e a sigla dos seus respectivos departamento

*Problema*

Sortear aleatoariamente os M avaliados entre os N avaliadores

*Restrição*

Um avaliador não pode avaliar um membro do seu próprio departamento

```python
import pandas as pd
import numpy as np
import random
```

```python
avaliadores = pd.read_csv('data/avaliadores.csv')
avaliados = pd.read_csv('data/avaliados.csv')
```

```python
avaliadores
```

```python
import pandas as pd
import random

# Criar listas para armazenar os registros do DataFrame "sorteio"
sorteio_records = []

# Função para escolher aleatoriamente avaliados com setor diferente
def escolher_avaliados(avaliador_setor, avaliados):
    avaliados_filtrados = avaliados[avaliados['Setor'] != avaliador_setor]
    if len(avaliados_filtrados) >= 2:
        return random.sample(list(avaliados_filtrados.iterrows()), 11)
    return None

# Iterar pelos registros do DataFrame "avaliadores"
for _, row_avaliador in avaliadores.iterrows():
    name_avaliador = row_avaliador['Name']
    setor_avaliador = row_avaliador['Setor']

    # Escolher aleatoriamente 15 avaliados com setor diferente
    avaliados_escolhidos = escolher_avaliados(setor_avaliador, avaliados)

    if avaliados_escolhidos:
        for index, row_avaliado in avaliados_escolhidos:
            sorteio_records.append({'Avaliador_Name': name_avaliador,
                                    'Avaliador_Setor': setor_avaliador,
                                    'Avaliado_Name': row_avaliado['Name'],
                                    'Avaliado_Setor': row_avaliado['Setor']})
            # Remover o avaliado escolhido para evitar repetições
            avaliados.drop(index, inplace=True)

# Criar o DataFrame "sorteio" a partir dos registros coletados
sorteio = pd.DataFrame(sorteio_records)

# Mostrar o DataFrame "sorteio" resultante
pd.set_option('display.max_rows', 165)
sorteio
```

```python
sorteio.to_csv('~/sorteio.csv', index=False)
```
