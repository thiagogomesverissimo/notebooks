```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('data.csv', index_col=0)
df
```

```python
# barplot com raça, gênero e cotista
df[['raça', 'gênero', 'cotista']]
```

```python
df["gênero-raça"] = df["gênero"] + ' ' + df["raça"]
```

```python
df['Forma de ingresso'] = df['cotista']
```

```python
# Exemplo:

import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="darkgrid")
titanic = sns.load_dataset("titanic")
ax = sns.countplot(x="class", hue="who", data=titanic)
```

```python
titanic
```

```python
ax = sns.countplot(x="gênero", hue="cotista", data=df)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.savefig('output/luiz.png')
plt.show()
```

```python
ax = sns.countplot(x="gênero-raça", hue="cotista", data=df)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.savefig('output/luiz2.png')
plt.show()
```

```python
import seaborn as sns
    
df = sns.load_dataset('titanic')
    
ax = sns.histplot(
   x = df['class'], 
   hue=df['survived'], 
   multiple="dodge",
   stat = 'percent',
   shrink = 0.5, 
   common_norm=False
)
```

```python
import seaborn as sns
    
ax = sns.histplot(
   x = df['gênero-raça'], 
   hue=df['Forma de ingresso'], 
   multiple="dodge",
   stat = 'percent',
   shrink = 0.5, 
   common_norm=False
)
ax.set(ylabel='Porcentagem')

ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.savefig('output/luiz3.png')
plt.show()
```
