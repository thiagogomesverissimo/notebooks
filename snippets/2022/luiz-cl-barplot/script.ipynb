```python
ax = sns.countplot(x="gênero", hue="cotista", data=df)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.savefig('output/luiz.png')
plt.show()
```
