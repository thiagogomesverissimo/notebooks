```python
import re
with open("data/maritaca.html", "r") as file:
    maritaca = file.read()

pattern="""
(https://api.spreaker.com)  # host
(.+?)
([&])  # host
"""

lista = re.findall(pattern,maritaca,re.VERBOSE)
lista
```

```python
output = []
for i in lista:
    output.append(i[0] + i[1])
output = list(set(output))
print(output)
```

```python
f = open("/tmp/maritaca.txt", "w")
f.write('\n'.join([str(elem) for elem in output]))
f.close()
```
