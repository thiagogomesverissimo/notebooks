```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd

from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.cluster.hierarchy import fcluster
from scipy.cluster.hierarchy import cophenet
from scipy.spatial.distance import pdist
```

```python
data = pd.read_csv('data/cluster.csv')
data.head()
```

```python
plt.scatter(data['x'], data['y'])
plt.axis('equal')
plt.show()
```

```python
# método do vizinho mais próximo
# métodos: complete, average e ward
LinkageSE = linkage(data, method='single', metric='euclidean', optimal_ordering=False)
dendrogram(LinkageSE,leaf_rotation=0.0,leaf_font_size=9.0)
plt.show()
```

```python
# distância cophenetic
SEc, SEc_dists = cophenet(LinkageSE, pdist(data))
print(SEc)
```

```python
# pintando as bolinhas agrupadas
SEclusters = fcluster(LinkageSE, t=1.0, criterion='distance')

#print(SEclusters)

plt.scatter(data['x'], data['y'],c=SEclusters)
plt.axis('equal')
plt.show()
```
