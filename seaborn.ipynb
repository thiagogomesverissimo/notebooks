```python
# pip3 install seaborn
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = sns.load_dataset("tips")
# Como as outras vari√°veis impactam na coluna "tip"?
data.head()
```

```python
data.describe()
```

```python
plot1 = sns.displot(data['total_bill'],kde=True)
plt.show()
```

```python
sns.displot(data['tip'],kde=False)
plt.show()
```

```python
sns.boxplot(x=data['total_bill']).set_title("boxplot da coluna total_bill")
plt.show()
```

```python
sns.boxplot(x=data['tip']).set_title("boxplot da coluna tip")
plt.show()
```

```python
sns.boxplot(x=data['total_bill'])
sns.boxplot(x=data['tip']).set_title("boxplot das colunas tip e total")
plt.show()
```

```python
# facets, y √© a vari√°vel categ√≥rica
sns.boxplot(x=data['tip'],y=data['smoker'])
plt.show()
```

```python
# facets, y √© a vari√°vel categ√≥rica
sns.boxplot(x=data['tip'],y=data['smoker'])
g = sns.FacetGrid(data,row="smoker")
g = g.map(plt.hist, "tip")
plt.show()
```

```python
# facets, y √© a vari√°vel categ√≥rica
sns.boxplot(x=data['tip'],y=data['time'])
plt.show()
```

```python
# facets, y √© a vari√°vel categ√≥rica
sns.boxplot(x=data['tip'],y=data['time'])
g = sns.FacetGrid(data,row="time")
g = g.map(plt.hist, "tip")
plt.show()
```

```python
# facets, y √© a vari√°vel categ√≥rica
sns.boxplot(x=data['tip'],y=data['day'])
plt.show()
```

```python
# facets, y √© a vari√°vel categ√≥rica
sns.boxplot(x=data['tip'],y=data['day']).set_title("Boxplot por dia da semana")
g = sns.FacetGrid(data,row="day")
g = g.map(plt.hist, "tip")
plt.show()
```

```python
sns.lmplot(x='tip', y='total_bill', data=data, fit_reg=False)   # Color by evolution stage
plt.show()
```

```python
sns.lmplot(x='tip', y='total_bill', data=data, fit_reg=True)
plt.show()
```

```python
sns.lmplot(x='tip', y='total_bill', data=data, fit_reg=False, hue='smoker')
plt.show()
```

```python
sns.countplot(x='smoker', data=data)
plt.xticks(rotation=-45)
plt.show()
```
